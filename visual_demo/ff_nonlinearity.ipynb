### Goal of this file is to demonstrate that feedforward layer in a transformer introduced non-linearity to the signal

```python
import torch
import torch.nn as nn
```

```python
import matplotlib.pyplot as plt
def plot_tensor(y:torch.Tensor, x:torch.Tensor):
    plt.plot(x.detach().numpy(),y.detach().numpy())
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot of x vs y')
    # y_min = max(y.min().item(), x.min().item())
    # y_max = min(y.max().item(), x.max().item())
    # x_min = max(x.min().item(), y.min().item())
    # x_max = min(x.max().item(), y.max().item())
    # margin = 0.5
    # plt.xlim(x_min-margin, x_max+margin)
    # plt.ylim(y_min-margin, y_max+margin)
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.show()
```

```python
x = torch.tensor([x for x in range(-1000,1000+1)], dtype=torch.float32)
```

```python
higher_dim = 15

ff = nn.Sequential(
    nn.Linear(1, higher_dim),
    nn.ReLU(),
    nn.Linear(higher_dim, 1)
)

def f(x):
    return ff(x.unsqueeze(-1)).squeeze(-1)    

plot_tensor(f(x), x)
```
